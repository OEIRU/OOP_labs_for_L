```python
import pandas as pd
import matplotlib.pyplot as plt
#    
file1 = 'test1/fs_empirical.txt' 
file2 = 'test1/fs_theoretical.txt' 
file3 = 'test1/xs.txt'  
#    DataFrame
data1 = pd.read_csv(file1, header=None)
data2 = pd.read_csv(file2, header=None)
data3 = pd.read_csv(file3, header=None)
#     
window_size = 10  #     
smoothed_data1 = data1.rolling(window=window_size).mean()
smoothed_data2 = data2.rolling(window=window_size).mean()
smoothed_data3 = data3.rolling(window=window_size).mean()
#  
plt.figure(figsize=(10, 6))
plt.plot(smoothed_data1, label=' 1 ()', marker='o')
plt.plot(smoothed_data2, label=' 2 ()', marker='x')
#plt.plot(smoothed_data3, label=' 3 ()', marker='s')
plt.title(' 1,  , n = 1000, v = 5')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.grid()
plt.show()
```
