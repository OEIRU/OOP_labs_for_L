```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def huber_distribution(x, v, K, scale, shift):
    #     
    if np.abs((x - shift) / scale) <= v:
        return (1 / (np.sqrt(2 * np.pi) * K)) * (1 / scale) * np.exp(-((x - shift)**2) / (2 * scale**2))
    else:
        return (1 / (np.sqrt(2 * np.pi) * K)) * (1 / scale) * np.exp((v**2 / 2) - v * np.abs((x - shift) / scale))

#   
v = 5.0
K = 1.0
scale = 1.0
shift = 0.0

#   x
x_values = np.linspace(-5, 5, 10000)
y_values = [huber_distribution(x, v, K, scale, shift) for x in x_values]

#  
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=' ', color='blue')
plt.title('     ')
plt.xlabel('x')
plt.ylabel(' ')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 
v = 5  #   
scale = 1  # 
shift = 0  # 
n = 1000  #   

#    ( )
data = np.random.normal(shift, scale, n)

#   
def huber_rvs(v, loc=0, scale=1, size=1):
    """       ."""
    normal_data = np.random.normal(loc, scale, size)
    absolute_data = np.abs(normal_data)
    return np.where(absolute_data <= v, normal_data, v * np.sign(normal_data) * (absolute_data - v))

huber_data = huber_rvs(v, loc=shift, scale=scale, size=n)

#  
plt.figure(figsize=(8, 6))
    
#  
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Empirical Distribution')

#   
#           
x_values = np.linspace(min(data), max(data), 1000)
huber_pdf = norm.pdf(x_values, loc=shift, scale=scale)  #    

plt.plot(x_values, huber_pdf, color='b', label='Huber Distribution', linewidth=2)

#  
plt.title('Empirical and Huber Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

plt.show()
```
