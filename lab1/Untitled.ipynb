```python
import matplotlib.pyplot as plt
import numpy as np

# Загрузка данных из файлов
xs = np.loadtxt('tests/data1/xs.txt')
fs_theoretical = np.loadtxt('tests/data1/fs_theoretical.txt')
fs_empirical = np.loadtxt('tests/data1/fs_empirical.txt')

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(xs, fs_theoretical, label='Теоретическая плотность', color='blue')
plt.plot(xs, fs_empirical, label='Эмпирическая плотность', color='red', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('Плотность')
plt.title('Тест №1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Параметры распределения Хьюбера
v = 2
scale = 1
shift = 0
n = 50000  # Размер выборки

# Генерация случайных значений из распределения Хьюбера
# Для простоты используем нормальное распределение, так как распределение Хьюбера
# с v = 2 близко к нормальному.
np.random.seed(42)  # Для воспроизводимости
data = np.random.normal(loc=shift, scale=scale, size=n)

# Теоретическая плотность распределения Хьюбера
def huber_density(x, v, scale, shift):
    return np.exp(-0.5 * np.abs((x - shift) / scale)**v) / (scale * np.sqrt(2 * np.pi))

# Создание сетки значений x для построения графиков
x_grid = np.linspace(-5, 5, 1000)  # Диапазон от -5 до 5 с 1000 точками
theoretical_density = huber_density(x_grid, v, scale, shift)

# Построение гистограммы для эмпирического распределения
plt.figure(figsize=(10, 6))
plt.hist(data, bins=50, density=True, alpha=0.6, color='g', label='Эмпирическое распределение')

# Построение теоретической плотности
plt.plot(x_grid, theoretical_density, 'r-', lw=2, label='Теоретическое распределение Хьюбера')

# Настройка графика
plt.title('Сравнение теоретического и эмпирического распределений Хьюбера')
plt.xlabel('x')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True)
plt.show()
```
