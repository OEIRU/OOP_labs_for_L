```python
import matplotlib.pyplot as plt
import numpy as np

#    
xs = np.loadtxt('tests/data1/xs.txt')
fs_theoretical = np.loadtxt('tests/data1/fs_theoretical.txt')
fs_empirical = np.loadtxt('tests/data1/fs_empirical.txt')

#  
plt.figure(figsize=(10, 6))
plt.plot(xs, fs_theoretical, label=' ', color='blue')
plt.plot(xs, fs_empirical, label=' ', color='red', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('')
plt.title(' 1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#   
v = 2
scale = 1
shift = 0
n = 50000  #  

#      
#     ,    
#  v = 2   .
np.random.seed(42)  #  
data = np.random.normal(loc=shift, scale=scale, size=n)

#    
def huber_density(x, v, scale, shift):
    return np.exp(-0.5 * np.abs((x - shift) / scale)**v) / (scale * np.sqrt(2 * np.pi))

#    x   
x_grid = np.linspace(-5, 5, 1000)  #   -5  5  1000 
theoretical_density = huber_density(x_grid, v, scale, shift)

#     
plt.figure(figsize=(10, 6))
plt.hist(data, bins=50, density=True, alpha=0.6, color='g', label=' ')

#   
plt.plot(x_grid, theoretical_density, 'r-', lw=2, label='  ')

#  
plt.title('     ')
plt.xlabel('x')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.show()
```
