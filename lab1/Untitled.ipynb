```python
import pandas as pd
import matplotlib.pyplot as plt
# Чтение данных из файлов
file1 = 'data/fs_empirical.txt' 
file2 = 'data/fs_theoretical.txt' 
file3 = 'data/xs.txt'  
# Чтение данных в DataFrame
data1 = pd.read_csv(file1, header=None)
data2 = pd.read_csv(file2, header=None)
data3 = pd.read_csv(file3, header=None)
# Применение скользящего среднего для сглаживания
window_size = 1  # Размер окна для скользящего среднего
smoothed_data1 = data1.rolling(window=window_size).mean()
smoothed_data2 = data2.rolling(window=window_size).mean()
smoothed_data3 = data3.rolling(window=window_size).mean()
# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(smoothed_data1, label='Файл 1 (сглаженный)', marker='o')
plt.plot(smoothed_data2, label='Файл 2 (сглаженный)', marker='x')
#plt.plot(smoothed_data3, label='Файл 3 (сглаженный)', marker='s')
plt.title('Тест 1, Стандартное распределение, n = 1000, v = 5')
plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def huber_distribution(x, v, K, scale, shift):
    # Функция плотности вероятности распределения Хубера
    if np.abs((x - shift) / scale) <= v:
        return (1 / (np.sqrt(2 * np.pi) * K)) * (1 / scale) * np.exp(-((x - shift)**2) / (2 * scale**2))
    else:
        return (1 / (np.sqrt(2 * np.pi) * K)) * (1 / scale) * np.exp((v**2 / 2) - v * np.abs((x - shift) / scale))

# Параметры распределения Хубера
v = 5.0
K = 1.0
scale = 1.0
shift = 0.0

# Генерация значений x
x_values = np.linspace(-5, 5, 10000)
y_values = [huber_distribution(x, v, K, scale, shift) for x in x_values]

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Распределение Хубера', color='blue')
plt.title('График функции плотности вероятности распределения Хубера')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Параметры
v = 5  # Параметр распределения Хьюберта
scale = 1  # Масштаб
shift = 0  # Смещение
n = 1000  # Количество сгенерированных точек

# Генерация эмперического распределения (нормальное распределение)
data = np.random.normal(shift, scale, n)

# Генерация распределения Хьюберта
def huber_rvs(v, loc=0, scale=1, size=1):
    """Функция для генерации случайных величин из распределения Хьюберта."""
    normal_data = np.random.normal(loc, scale, size)
    absolute_data = np.abs(normal_data)
    return np.where(absolute_data <= v, normal_data, v * np.sign(normal_data) * (absolute_data - v))

huber_data = huber_rvs(v, loc=shift, scale=scale, size=n)

# Построение графика
plt.figure(figsize=(8, 6))
    
# Эмперическое распределение
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Empirical Distribution')

# Теоретическое распределение Хьюберта
# Для построения линии используем ядро плотности или эмпирическое распределение для Хьюберта
x_values = np.linspace(min(data), max(data), 1000)
huber_pdf = norm.pdf(x_values, loc=shift, scale=scale)  # Теоретическая плотность нормального распределения

plt.plot(x_values, huber_pdf, color='b', label='Huber Distribution', linewidth=2)

# Настройки графика
plt.title('Empirical and Huber Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

plt.show()
```
